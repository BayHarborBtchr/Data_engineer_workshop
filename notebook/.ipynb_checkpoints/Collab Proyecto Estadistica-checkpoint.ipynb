```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url_alaska = 'https://github.com/jpgomezv/Proyecto1Estadistica/raw/main/alaska_students_sample.csv'
url_newyork = 'https://github.com/jpgomezv/Proyecto1Estadistica/raw/main/newyork_students_sample.csv'
url_ohio = 'https://github.com/jpgomezv/Proyecto1Estadistica/raw/main/ohio_students_sample.csv'
url_texas = 'https://github.com/jpgomezv/Proyecto1Estadistica/raw/main/texas_students_sample.csv'

alaskadf = pd.read_csv(url_alaska, encoding='latin1')
newyorkdf = pd.read_csv(url_newyork, encoding='latin1')
ohiodf = pd.read_csv(url_ohio, encoding='latin1')
texasdf = pd.read_csv(url_texas, encoding='latin1')

newyorkdf = newyorkdf.copy()
ohiodf = ohiodf.copy()
texasdf = texasdf.copy()
alaskadf = alaskadf.copy()

raw_dfmixed = pd.concat([newyorkdf, ohiodf, texasdf, alaskadf], ignore_index=True)

# Los archivos CSV de muestras de estudiantes están almacenados en GitHub para un acceso constante.
# Se crean DataFrames a partir de estos archivos utilizando la codificación 'latin1'.
```

```python
irrelevant_columns = [
    'Country',
    'ClassGrade',
    'Left_Footlength_cm',
    'Longer_foot',
    'Index_Fingerlength_mm',
    'Ring_Fingerlength_mm',
    'Longer_Finger_Lefthand',
    'Favorite_Season',
    'Allergies',
    'Vegetarian',
    'Favorite_Food',
    'Home_Internet_Access',
    'Communication_With_Friends',
    'Text_Messages_Sent_Yesterday',
    'Text_Messages_Received_Yesterday',
    'Hanging_Out_With_Friends_Hours',
    'Talking_On_Phone_Hours',
    'Doing_Things_With_Family_Hours',
    'Outdoor_Activities_Hours',
    'Video_Games_Hours',
    'Social_Websites_Hours',
    'Texting_Messaging_Hours',
    'Travel_to_School',
    'Computer_Use_Hours',
    'Watching_TV_Hours',
    'Schoolwork_Pressure',
    'Favorite_Music',
    'Superpower',
    'Preferred_Status',
    'Role_Model_Type',
    'Charity_Donation'
]

newdfmixed = raw_dfmixed.drop(columns=irrelevant_columns)
print(newdfmixed.head())
```

```python
#cantidad de filas inicial, claramente 2000 xD
print(len(newdfmixed))

#info gral sobre el df (tipos de datos, valores nulos, etc)
print(newdfmixed.info())

# descripción estadística de las columnas numéricas
print(newdfmixed.describe())
```

```python
numeric_columns = [
    'DataYear',
    'Ageyears',
    'Height_cm',
    'Footlength_cm',
    'Armspan_cm',
    'Languages_spoken',
    'Travel_time_to_School',
    'Reaction_time',
    'Score_in_memory_game',
    'Importance_reducing_pollution',
    'Importance_conserving_water',
    'Importance_saving_energy',
    'Importance_owning_computer',
    'Importance_Internet_access',
    'Sleep_Hours_Schoolnight',
    'Sleep_Hours_Non_Schoolnight',
    'Home_Occupants',
    'Doing_Homework_Hours',
    'Paid_Work_Hours',
    'Work_At_Home_Hours'
]

for col in numeric_columns:
    newdfmixed[col] = pd.to_numeric(newdfmixed[col], errors='coerce')

# Verificar los cambios
print(newdfmixed.info())
print(len(newdfmixed))
```

```python
print(newdfmixed.isnull().sum())
print(len(newdfmixed))
```

```python
# Elimina las filas con valores nulos en cualquier columna
dfmixed = newdfmixed.dropna()
print(len(dfmixed))
```

```python
# Definir los límites solo para las columnas de "Importance"
limites_importance = {
    "Importance_reducing_pollution": (0, 1000),
    "Importance_recycling_rubbish": (0, 1000),
    "Importance_conserving_water": (0, 1000),
    "Importance_saving_energy": (0, 1000),
    "Importance_owning_computer": (0, 1000),
    "Importance_Internet_access": (0, 1000)
}

# Aplicar los límites
for col, (min_val, max_val) in limites_importance.items():
    dfmixed[col] = pd.to_numeric(dfmixed[col], errors='coerce')  # Convertir a numérico
    dfmixed = dfmixed[(dfmixed[col] >= min_val) & (dfmixed[col] <= max_val)]

# Mostrar el DataFrame resultante
print(dfmixed.head())
print(len(dfmixed))
```

```python
#guardamos datos limpios
finaldf = dfmixed.copy()
finaldf.to_csv('bd_depurada.csv', index=False, encoding='latin1')
```

# 7. Construya una Ojiva para la variable “Tiempo que dedican los estudiantespara ayudar a los quehaceres del Hogar” y a partir de esta grafica responda las siguientes preguntas:

```python
import numpy as np
import matplotlib.pyplot as plt

work_at_home_hours = dfmixed['Work_At_Home_Hours'].dropna()

# Filtrar valores extremos (por ejemplo, eliminamos datos mayores al percentil 95)
upper_limit = work_at_home_hours.quantile(0.95)
filtered_hours = work_at_home_hours[work_at_home_hours <= upper_limit]

# Crear histograma y ojiva base
count, bins, _ = plt.hist(filtered_hours, bins=10, alpha=0.5, color='blue', edgecolor='black')
cumulative = np.cumsum(count)
cumulative_percentage = cumulative / cumulative[-1] * 100
bins_mid = [(bins[i] + bins[i+1]) / 2 for i in range(len(bins)-1)]

def plot_histogram_and_ojiva(title, x_label, y_label, line_value=None, line_text=None, line_color='green'):
    plt.figure(figsize=(6, 6))
    plt.hist(filtered_hours, bins=10, alpha=0.5, color='blue', edgecolor='black')
    plt.plot(bins_mid, cumulative_percentage, marker='o', color='red', label='Ojiva (Porcentaje)')

    if line_value is not None:
        plt.axvline(line_value, color=line_color, linestyle='--', label=f'{line_text}')
        plt.text(line_value, max(cumulative_percentage), f'{line_text}', color=line_color, verticalalignment='center')

    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend()
    plt.grid(True)
    plt.show()

# a) Porcentaje de estudiantes que dedican mínimo 15 horas a la semana
students_min_15h = (filtered_hours >= 15).sum()
percent_min_15h = (students_min_15h / len(filtered_hours)) * 100
plot_histogram_and_ojiva(
    title='Gráfica de Ojiva para Horas Trabajadas desde Casa',
    x_label='Horas trabajadas desde casa',
    y_label='Frecuencia acumulada (%)',
    line_value=15,
    line_text=f'Mínimo 15 horas ({percent_min_15h:.2f}%)',
    line_color='green'
)

# b) Porcentaje de estudiantes que dedican máximo 5 horas a la semana
students_max_5h = (filtered_hours <= 5).sum()
percent_max_5h = (students_max_5h / len(filtered_hours)) * 100
plot_histogram_and_ojiva(
    title='Gráfica de Ojiva para Horas Trabajadas desde Casa',
    x_label='Horas trabajadas desde casa',
    y_label='Frecuencia acumulada (%)',
    line_value=5,
    line_text=f'Máximo 5 horas ({percent_max_5h:.2f}%)',
    line_color='purple'
)

# c) Máximo número de horas que el 15 % de los estudiantes ayudan en sus casas
hours_15_percentile = filtered_hours.quantile(0.15)
plot_histogram_and_ojiva(
    title='Gráfica de Ojiva para Horas Trabajadas desde Casa',
    x_label='Horas trabajadas desde casa',
    y_label='Frecuencia acumulada (%)',
    line_value=hours_15_percentile,
    line_text=f'15% Percentil ({hours_15_percentile:.2f} horas)',
    line_color='orange'
)

# d) Porcentaje de estudiantes que ayudan en sus casas al menos 5 horas a la semana
students_at_least_5h = (filtered_hours >= 5).sum()
percent_at_least_5h = (students_at_least_5h / len(filtered_hours)) * 100
plot_histogram_and_ojiva(
    title='Gráfica de Ojiva para Horas Trabajadas desde Casa',
    x_label='Horas trabajadas desde casa',
    y_label='Frecuencia acumulada (%)',
    line_value=5,
    line_text=f'Al menos 5 horas ({percent_at_least_5h:.2f}%)',
    line_color='blue'
)

# e) Mínimo número de horas que el 35 % de los estudiantes ayudan en sus casas
hours_35_percentile = filtered_hours.quantile(0.65)
plot_histogram_and_ojiva(
    title='Gráfica de Ojiva para Horas Trabajadas desde Casa',
    x_label='Horas trabajadas desde casa',
    y_label='Frecuencia acumulada (%)',
    line_value=hours_35_percentile,
    line_text=f'35% Percentil ({hours_35_percentile:.2f} horas)',
    line_color='brown'
)

# f) Intervalo que contiene la moda
mode_hours = filtered_hours.mode()[0]
plot_histogram_and_ojiva(
    title='Gráfica de Ojiva para Horas Trabajadas desde Casa',
    x_label='Horas trabajadas desde casa',
    y_label='Frecuencia acumulada (%)',
    line_value=mode_hours,
    line_text=f'Moda ({mode_hours} horas)',
    line_color='magenta'
)

# g) Promedio y desviación estándar de las horas ayudando en casa
average_hours = filtered_hours.mean()
std_dev_hours = filtered_hours.std()
plot_histogram_and_ojiva(
    title='Gráfica de Ojiva para Horas Trabajadas desde Casa',
    x_label='Horas trabajadas desde casa',
    y_label='Frecuencia acumulada (%)',
    line_value=average_hours,
    line_text=f'Promedio: {average_hours:.2f} horas\nDesviación estándar: {std_dev_hours:.2f}',
    line_color='cyan'
)
```

```python
import pandas as pd

# Cargar los datos desde el archivo CSV
data = pd.read_csv('bd_depurada.csv', encoding='latin1')

# Filtrar los datos para mantener solo las horas de sueño en un rango razonable (por ejemplo, 0 a 24 horas)
filtered_data = data[(data['Sleep_Hours_Non_Schoolnight'] >= 0) & (data['Sleep_Hours_Non_Schoolnight'] <= 24)]

# a) ¿Qué porcentaje de estudiantes de la muestra son mujeres?
female_percentage = (filtered_data['Gender'] == 'Female').mean() * 100

# b) ¿Cuántas horas de sueño en promedio suelen dormir cuando no tiene clases al día siguiente los estudiantes de la muestra?
average_sleep_hours = filtered_data['Sleep_Hours_Non_Schoolnight'].mean()

# c) ¿En términos generales qué género duerme más tiempo?
average_sleep_by_gender = filtered_data.groupby('Gender')['Sleep_Hours_Non_Schoolnight'].mean()
gender_more_sleep = average_sleep_by_gender.idxmax()

# d) ¿En términos generales qué género tiene los tiempos más homogéneos?
sleep_std_by_gender = filtered_data.groupby('Gender')['Sleep_Hours_Non_Schoolnight'].std()
gender_more_homogeneous = sleep_std_by_gender.idxmin()

# e) ¿En términos generales qué género tiene los tiempos más simétricos?
sleep_skew_by_gender = filtered_data.groupby('Gender')['Sleep_Hours_Non_Schoolnight'].skew()
gender_more_symmetrical = sleep_skew_by_gender.abs().idxmin()

# f) ¿Cuál género tiene un histograma más puntiagudo? ¿Cuál género tiene un histograma más aplanado o chato?
sleep_kurtosis_by_gender = filtered_data.groupby('Gender')['Sleep_Hours_Non_Schoolnight'].apply(pd.Series.kurt)
gender_more_pointed = sleep_kurtosis_by_gender.idxmax()
gender_more_flat = sleep_kurtosis_by_gender.idxmin()

# Compilar los resultados en un formato legible
results = {
    "a) ¿Qué porcentaje de estudiantes de la muestra son mujeres?": f"{female_percentage:.2f}%",
    "b) ¿Cuántas horas de sueño en promedio suelen dormir cuando no tiene clases al día siguiente los estudiantes de la muestra?": f"{average_sleep_hours:.2f} horas",
    "c) ¿En términos generales qué género duerme más tiempo?": gender_more_sleep,
    "d) ¿En términos generales qué género tiene los tiempos más homogéneos?": gender_more_homogeneous,
    "e) ¿En términos generales qué género tiene los tiempos más simétricos?": gender_more_symmetrical,
    "f) ¿Cuál género tiene un histograma más puntiagudo? ¿Cuál género tiene un histograma más aplanado o chato?": f"Más puntiagudo: {gender_more_pointed}, Más aplanado: {gender_more_flat}"
}

# Mostrar los resultados
for question, answer in results.items():
    print(f"{question}\nRespuesta: {answer}\n")
```

```python
import pandas as pd

# Cargar los datos (guardarlos en un archivo CSV)
finaldf.to_csv('bd_depurada.csv', index=False, encoding='latin1')

# Cargar nuevamente el CSV que acabas de crear (para asegurarte de que todo se guardó correctamente)
data = pd.read_csv('bd_depurada.csv', encoding='latin1')

# Crear una tabla cruzada entre "Materia preferida en el Colegio" y "Género"
crosstab_subject_gender = pd.crosstab(data['Favorite_School_Subject'], data['Gender'])

# Definir todas las variables necesarias para las respuestas
male_history = crosstab_subject_gender.loc['History', 'Male']
percentage_male = (data['Gender'] == 'Male').mean() * 100
percentage_music = (data['Favorite_School_Subject'] == 'Music').mean() * 100
other_activities_women = (data[(data['Gender'] == 'Female') & (data['Favorite_School_Subject'] == 'Other')].shape[0] /
                          data[data['Gender'] == 'Female'].shape[0]) * 100
percentage_male_science = (data[(data['Favorite_School_Subject'] == 'Science') & (data['Gender'] == 'Male')].shape[0] /
                           data[data['Favorite_School_Subject'] == 'Science'].shape[0]) * 100
percentage_female_other = (data[(data['Gender'] == 'Female') & (data['Favorite_School_Subject'] == 'Other')].shape[0] /
                           data.shape[0]) * 100
percentage_female_or_music = (data[(data['Gender'] == 'Female') | (data['Favorite_School_Subject'] == 'Music')].shape[0] /
                              data.shape[0]) * 100
percentage_female_music_or_science = (data[(data['Gender'] == 'Female') &
                                           ((data['Favorite_School_Subject'] == 'Music') |
                                            (data['Favorite_School_Subject'] == 'Science'))].shape[0] /
                                      data[data['Gender'] == 'Female'].shape[0]) * 100

# Questions and answers
answers = {
    "a) ¿A cuántos estudiantes hombres les gusta la historia?": male_history,
    "b) ¿Qué porcentaje de estudiantes son hombres?": f"{percentage_male:.2f}%",
    "c) ¿Qué porcentaje de estudiantes prefieren música?": f"{percentage_music:.2f}%",
    "d) ¿De las mujeres, qué porcentaje de estudiantes prefieren otras actividades?": f"{other_activities_women:.2f}%",
    "e) ¿De los estudiantes que prefieren Ciencias, qué porcentaje son hombres?": f"{percentage_male_science:.2f}%",
    "f) ¿Qué porcentaje son mujeres y prefieren otras actividades?": f"{percentage_female_other:.2f}%",
    "g) ¿Qué porcentaje son mujeres o prefieren música?": f"{percentage_female_or_music:.2f}%",
    "h) ¿De las mujeres, qué porcentaje le gusta la música o las Ciencias?": f"{percentage_female_music_or_science:.2f}%"
}

# Imprimir las respuestas
for question, answer in answers.items():
    print(f"{question}\nRespuesta: {answer}\n")

# Análisis
print("Análisis:")
print("Los datos muestran que la historia es una materia preferida principalmente por los hombres (119),")
print("mientras que un porcentaje significativo de estudiantes son hombres (49.40%).")
print("La música es una materia menos preferida en general (7.02%), y es preferida por un porcentaje menor de estudiantes.")
print("En cuanto a las mujeres, una menor proporción (8.68%) prefiere otras actividades fuera de las materias principales,")
print("y entre aquellos que prefieren Ciencias, casi la mitad son hombres (44.69%).")
print("Además, un 4.39% del total de estudiantes son mujeres que prefieren otras actividades, mientras que un 53.45% son mujeres o prefieren música.")
print("Finalmente, un 22.32% de las mujeres muestra preferencia por la música o las Ciencias, lo que indica un interés moderado en estas materias entre las estudiantes.")
```
